### LIKE: De los productos, lista solo aquellos que sabemos que son "Corvette" por nombre, y ordénalos por cantidad de stock ascendente.

LIKE nos permite buscar patrones:

* % significa cualquier carácter, por lo que n% buscaría cualquier palabra que comience con "n".
* _ significa que debe haber cualquier carácter, por lo que l__ encontraría palabras con tres letras, comenzando con "l".
Estos comodines pueden combinarse para refinar aún más la búsqueda.


````python
q= '''
SELECT productName,quantityInStock
FROM products
where productname LIKE "%Corvette%" -- productos que contienen la palabra corvette(al final), %Corvett% productos que contienen la palabra corvette en cualquier parte
                                   -- Corvette% productos que contienen la palabra corvette al principio
'''                                   
pd.read_sql(q,con)
````

````python
q= '''
SELECT productName,quantityInStock
FROM products
where productname LIKE "_____Corvette%" -- el numero de guiones bajos es el numero de espacios/caracteres antes de corvette
                                  
'''                                   
pd.read_sql(q,con)
````

* strftime('%H', ExampleDate): Devuelve la hora.


### Promedio del límite de crédito por país para clientes con un límite > 0. Incluye el número de clientes que contribuyen a ese promedio. teniendo una media superior a 50.000


````python
q="""
select  country,count(customernumber),Cast(avg(creditlimit) as integer)
from customers
where creditlimit > 0
group by country
having avg(creditlimit) > 50000

"""
pd.read_sql(q,con)
````

 ### CAST(): Para convertir el formato de un campo, utilizamos la expresión CAST(campo AS tipo). El tipo puede ser INTEGER, FLOAT, char, date, datetime...

````python
q="""
select  country,count(customernumber),Cast(avg(creditlimit) as integer)
from customers
where creditlimit > 0
group by country
having avg(creditlimit) > 50000

"""
pd.read_sql(q,con)
````

La estructura de la función de ventana es la siguiente:

### Función de ranking

Las tres funciones asignarán un número a nuestra fila, pero hay diferencias entre ellas.

* ROW_NUMBER(): Asignará números continuos, y ningún número se repetirá en diferentes filas dentro de la misma partición.
* RANK(): Asignará el mismo número en caso de que haya filas con los mismos valores de ordenamiento y partición. Si esto ocurre, creará brechas dentro de las filas de la misma partición.
* DENSE_RANK(): Asignará el mismo número en caso de que haya filas con los mismos valores de ordenamiento y partición, pero no tendrá brechas dentro de la partición.

#### Crea un ranking de clientes con la mayoría de pedidos para cada país. Primero con una subconsulta:

RANK: Si dos clientes del mismo país tienen el mismo número de pedidos, tendrán el mismo número de rango, y el siguiente cliente considerará ambas filas con el mismo ranking.

DENSE_RANK: Si dos clientes del mismo país tienen la misma cantidad de pedidos, tendrán el mismo número de rango, y el siguiente cliente continuará con la secuencia.

ROW_NUMBER: Si dos clientes del mismo país tienen la misma cantidad de pedidos, tendrán clasificaciones diferentes y se ordenarán de forma aleatoria.
